```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#url = 'https://drive.google.com/file/d/12RUG1GqosS4Np15ZsGRJPDfQIkQddLEE/view?usp=drive_link'
# Leer el archivo CSV
data = pd.read_csv('Moocs_SUCIOS/Coursera_courses.csv')

# Extraer la columna deseada
columna_texto = data['name']

# Unir todos los textos en un solo string
texto_completo = ' '.join(columna_texto)

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_completo)

# Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
