```python
duplicates = df.duplicated()
```

```python
print(df[duplicates])
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df_cleaned.to_csv('UDEMY_cleaned.csv', index=False)
```

```python
# Paso 6: Visualizar la distribución de las variables numéricas utilizando gráficos de violín y cajas
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_cleaned, y='price')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cleaned, y='price')
plt.show()
```

```python
# Paso 7: Visualizar la relación entre variables numéricas utilizando un gráfico de dispersión y una matriz de dispersión
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_cleaned, x='price', y='num_subscribers')
plt.show()

plt.figure(figsize=(12, 10))
sns.pairplot(data=df_cleaned, vars=['num_subscribers', 'num_reviews', 'num_lectures'])
plt.show()
```

```python
correlation_matrix = df_cleaned.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Paso 9: Explorar las variables categóricas utilizando gráficos de barras y diagramas de dispersión
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='subject')
plt.show()

plt.figure(figsize=(10, 8))
sns.stripplot(data=df, x='subject', y='price')
plt.show()
```
