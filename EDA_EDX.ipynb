```python
# O
df['summary'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
df['n_enrolled'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
df['instructors'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
df['subtitles'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
df['course_description'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
df['course_syllabus'].fillna('sin dato', inplace=True)  # Rellenar valores faltantes con un valor específico
```

```python
# Paso 5: Resumen estadístico del DataFrame
print(df.describe())
```

```python
df.describe()
```

```python
duplicates = df.duplicated()
```

```python
print(df[duplicates])
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.head()
```

```python
df_cleaned.to_csv('EDX_cleaned.csv', index=False)
```

```python
# Paso 6: Visualizar la distribución de las variables numéricas utilizando gráficos de violín y cajas
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_cleaned, y='n_enrolled')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cleaned, y='n_enrolled')
plt.show()
```

```python
# Paso 7: Visualizar la relación entre variables numéricas utilizando un gráfico de dispersión y una matriz de dispersión
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_cleaned, x='columna_numerica1', y='columna_numerica2')
plt.show()

plt.figure(figsize=(12, 10))
sns.pairplot(data=df_cleaned, vars=['columna_numerica1', 'columna_numerica2', 'columna_numerica3'])
plt.show()
```

```python
# Paso 8: Calcular la matriz de correlación y visualizarla como un mapa de calor
correlation_matrix = df_cleaned.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Paso 9: Explorar las variables categóricas utilizando gráficos de barras y diagramas de dispersión
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='columna_categorica')
plt.show()

plt.figure(figsize=(10, 8))
sns.stripplot(data=df, x='columna_categorica', y='columna_numerica')
plt.show()
```
